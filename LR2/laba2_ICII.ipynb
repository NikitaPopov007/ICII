## 3. Отобразить список наименований средств, соответствующих двум любым документам с требованиями по сертификации

```python
dok = ['Соответствует требованиям документов: Профили защиты СОВ(cети четвертого класса защиты. ИТ.СОВ.С4.ПЗ), ЗБ, РД МЭ(2), РД НДВ(2)',  'Соответствует требованиям документов: ЗБ, РД НДВ(4), РД БИТ(ОУД2)'] 

ind1 = df.index[df['Наименования документов, требованиям которых соответствует средство'] == dok[0]].tolist()
ind2 = df.index[df['Наименования документов, требованиям которых соответствует средство'] == dok[1]].tolist()

index = np.hstack((ind1, ind2))
result = df.iloc[index]
result
```

## 4. Показать наименования тех средств, которые имеют просроченные сертификаты

Преобразовываем сразу два столбца в тип данных "datetime", так как в задании №5 это понадобится.

```python
df[['Срок\r\nдействия\r\nсертификата', 'Информация об окончании срока технической поддержки, полученная от заявителя']] = df[['Срок\r\nдействия\r\nсертификата', 'Информация об окончании срока технической поддержки, полученная от заявителя']].apply(pd.to_datetime, format='%d.%m.%Y', errors='coerce')
date = pd.Timestamp.today()
print("Текущая дата:", date.year, date.month, date.day)
```

```python
result = df[(df['Срок\r\nдействия\r\nсертификата'] < date)]
result
```

## 5. Показать наименования тех средств, которые имеют просроченные сертификаты, но при этом обязаны оказывать поддержку (столбец «Информация об окончании срока технической поддержки, полученная от заявителя»)

```python
result = df[(df['Срок\r\nдействия\r\nсертификата'] < date) & (df['Информация об окончании срока технической поддержки, полученная от заявителя'] > date)]
result
```
